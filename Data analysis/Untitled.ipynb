```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# ideal gaussian
xdata = np.linspace(-5,5,1000)

def gaussian(x, mu, sig):
    return (1/(np.sqrt(2*np.pi)*sig)) * np.exp(-0.5*((x-mu)/sig)**2)

ydata = gaussian(xdata, 0, 1)
# plt.plot(xdata, ydata)
```

```python
plt.plot(xdata, ydata)
# chop the gaussian
sigma_n = 2.2
ydata[(xdata < -sigma_n) | (xdata > sigma_n)]=0
plt.plot(xdata, ydata)
```

```python
np.sum(ydata)
```

```python
# gaussian fit
from scipy.optimize import curve_fit

p, pcov = curve_fit(gaussian, xdata, ydata)
print('Gaussian fit: mu='+str(p[0])+' sigma='+str(p[1]))
plt.plot(xdata, ydata)
plt.plot(xdata, gaussian(xdata, *p))
```
