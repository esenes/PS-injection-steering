```python
ctime1, emix_mean1, emix_std1, emiy_mean1, emiy_std1 = np.loadtxt('emi_nominal_RFon.txt').T
ctime2, emix_mean2, emix_std2, emiy_mean2, emiy_std2 = np.loadtxt('emi_nominal_RFoff.txt').T
ctime3, emix_mean3, emix_std3, emiy_mean3, emiy_std3 = np.loadtxt('emi_200urad_RFon.txt').T
ctime4, emix_mean4, emix_std4, emiy_mean4, emiy_std4 = np.loadtxt('emi_200urad_RFoff.txt').T
```

```python
fig, ax = plt.subplots(4,2,figsize=(10,15), sharex=True)

ax[0,0].errorbar(ctime1, emix_mean1, yerr=emix_std1, fmt='o',label='Nominal_RFon')
ax[1,0].errorbar(ctime1, emix_mean2, yerr=emix_std2, fmt='o',label='Nominal_RFoff')
ax[2,0].errorbar(ctime1, emix_mean3, yerr=emix_std3, fmt='o',label='200urad_RFon')
ax[3,0].errorbar(ctime1, emix_mean4, yerr=emix_std4, fmt='o',label='200urad_RFoff')



ax[0,1].errorbar(ctime1, emiy_mean1, yerr=emiy_std1, fmt='o',label='Nominal_RFon')
ax[1,1].errorbar(ctime1, emiy_mean2, yerr=emiy_std2, fmt='o',label='Nominal_RFoff')
ax[2,1].errorbar(ctime1, emiy_mean3, yerr=emiy_std3, fmt='o',label='200urad_RFon')
ax[3,1].errorbar(ctime1, emiy_mean4, yerr=emiy_std4, fmt='o',label='200urad_RFoff')

ax[3,0].set_xlabel('Cycle time (ms)')
# ax[0].set_ylabel('Norm. emittance (mm.mrad)')
ax[3,1].set_xlabel('Cycle time (ms)')

ax[0,0].set_title('X PLANE')
ax[0,1].set_title('Y PLANE')


# set axis to get rid of some bad shots
# x1,x2,y1,y2 = ax[0].axis()
# ax[0].axis((172,x2,y1,1.8))



# ax[0].legend(frameon=True)
# ax[1].legend(frameon=True)
fig.tight_layout()

fig.savefig('emix_y_along_cycle.eps', format='eps',dpi=100)
```
